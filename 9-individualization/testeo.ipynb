```python

text_path="D:/documentos/universidad/TESIS/parcellation/parcellation-master/8-parcellation/output/trac_atlas/right_bundle_labels.txt"
# Read data from the text file and create a dictionary
data_dict = {}
with open(text_path, 'r') as file:
    for line in file:
        key, value = line.strip().split(' ', 1)
        data_dict[int(key)] = value

# Print the dictionary
sorted_dict = dict(sorted(data_dict.items()))
for key, value in sorted_dict.items():
    print(key, value)
```
